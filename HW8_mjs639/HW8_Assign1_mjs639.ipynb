```python
import pylab as pl
import pandas as pd
import numpy as np
import os
import scipy.stats

%pylab inline
```

```python
df1 = pd.read_csv('https://data.cityofnewyork.us/api/views/f9bf-2cp4/rows.csv?accessType=DOWNLOAD')
df1 = df1.rename(columns={'SAT Critical Reading Avg. Score': 'READING AVG', 'SAT Math Avg. Score': 'MATH AVG',
                         'SAT Writing Avg. Score': 'WRITING AVG'})
```

```python
df1.head()
```

```python
df2 = pd.read_csv('https://data.cityofnewyork.us/api/views/n3p6-zve2/rows.csv?accessType=DOWNLOAD')
df2 = df2.drop(['building_code', 'phone_number', 'fax_number', 'grade_span_min', 
         'grade_span_max', 'expgrade_span_min', 'expgrade_span_max', 'priority08', 'priority09',
         'priority10', 'bus', 'subway', 'primary_address_line_1', 'city', 'state_code',
         'postcode', 'website', 'campus_name', 'school_type', 'overview_paragraph',
         'program_highlights', 'language_classes', 'advancedplacement_courses', 'online_ap_courses',
         'online_language_courses', 'extracurricular_activities', 'Community Board', 'Council District',      
         'priority03', 'priority04', 'priority05', 'priority01', 'priority02', 'priority06', 'priority07',
         'psal_sports_boys', 'psal_sports_girls', 'psal_sports_coed', 'school_sports', 'partner_cbo',      
         'partner_hospital', 'partner_highered', 'partner_cultural', 'partner_nonprofit', 'partner_corporate',
         'partner_financial', 'partner_other', 'end_time', 'se_services', 'ell_programs', 
         'school_accessibility_description', 'number_programs', 'addtl_info1', 'addtl_info2', 'start_time',
         'Census Tract', 'BIN', 'BBL', 'NTA', 'school_name'], axis=1, inplace=False)
df2 = df2.rename(columns={'dbn': 'DBN', 'Location 1': 'ADDRESS'})
```

```python
df2.head()
```

```python
df = pd.merge(df1, df2, on='DBN')
```

```python
df.head(20)
```

```python
df = df[df['READING AVG'] != 's']
df.head(20)
```

```python
df['NEW ADDRESS'] = df['ADDRESS'].str.split('\n', 1, expand=True)[1]
df['LOCATION'] = df['NEW ADDRESS'].str.split('\n', 1, expand=True)[1]
```

```python
df.head()
```

```python
dfCarto = df.drop(['DBN', 'ADDRESS', 'NEW ADDRESS', 'SCHOOL NAME', 'Num of SAT Test Takers', 'borough'], 
                  axis=1, inplace = False)
```

```python
dfCarto.head()
```

```python
#Convert values to Integers

reading = []
for i in dfCarto['READING AVG']:
    integer = int(i)
    reading.append(integer)

math = []
for i in dfCarto['MATH AVG']:
    integer = int(i)
    math.append(integer)
    
writing = []
for i in dfCarto['WRITING AVG']:
    integer = int(i)
    writing.append(integer)

    
dfCarto['READING AVG'] = reading
dfCarto['MATH AVG'] = math
dfCarto['WRITING AVG'] = writing


dfCarto['TOTAL'] = dfCarto['READING AVG'] + dfCarto['MATH AVG'] + dfCarto['WRITING AVG']
```

```python
dfCarto.head()
```

```python
#Generating Latitude and Longitude

location = []
for i in dfCarto['LOCATION']:
    newstring = i.replace("(", "")
    newstring2 = newstring.replace(")", "")
    location.append(newstring2)
    
dfCarto['LOCATION'] = location
dfCarto['latitude'] = dfCarto['LOCATION'].str.split(',', 1, expand=True)[0]
dfCarto['longitude'] = dfCarto['LOCATION'].str.split(',', 1, expand=True)[1]   
```

```python
dfCarto.head()
```

```python
dfCarto.to_csv('SATCartoData.csv')
```
