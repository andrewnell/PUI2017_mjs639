```python
from __future__ import print_function
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
import pylab as pl
%pylab inline

#Seed and Constant Variables
np.random.seed(10)
sigma = 30
mu = 11

#Initial Random Array
ReprRand = np.random.randn(2, 100)
ReprRand.shape

#50 Random Arrays
ReprRandAll = np.zeros((50, 2, 100))
for i in range(50):
        ReprRandAll[i] = sigma * np.random.randn(2,100) + (mu+(10*i))

#Plotting
fig = plt.figure(figsize = (10,8))
fig.suptitle('Random Reproducible Data', fontsize=16)
plt.xlabel('Random X Values, Seed 10')
plt.ylabel('Random Y Values, Seed 10')
for i in range(50):
    plt.plot(ReprRandAll[i][0], ReprRandAll[i][1], c=np.random.rand(3,1), linestyle="",marker="o")
plt.show() 
print('Fig 1: This figure is a plot of random reproducible data, with a seed value of 10')
```
