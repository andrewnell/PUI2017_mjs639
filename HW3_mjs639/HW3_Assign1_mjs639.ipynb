```python
from __future__ import print_function
import pylab as pl
import numpy as np
import seaborn
%pylab inline
```

## CONSTANTS AND UNIVERSAL VARIABLES

```python
np.random.seed(150)
mymean = 450
mySD = 37
mydict = {}
mysize = (np.random.rand(100) * 2000).astype(int)
print (mysize)
mysize.shape
```

## Chi-Square Distribution

```python
#Setting the seed, the dictionary key, and key variables
np.random.seed(50)
mydict['chisq'] = {} 
df = mymean

#Filling the dictionary
for n in mysize:
    mydict['chisq'][n] = np.random.chisquare(df, size = n)

#Saving the means in my dictionary
mydict['chisq']['means'] = {}

#Plotting the Means
axChisq_means = pl.figure(figsize=(15,8)).add_subplot(111)
for nn in mydict['chisq']:
    if not type(nn) == str:
        mydict['chisq']['means'][nn] = mydict['chisq'][nn].mean()
        axChisq_means.plot(nn, mydict['chisq']['means'][nn], 'o')
        axChisq_means.set_xlabel('sample size', fontsize=18)
        axChisq_means.set_ylabel('sample mean', fontsize=18)
        axChisq_means.set_title('Chi squared', fontsize=18)
        axChisq_means.plot([min(mysize), max(mysize)], [mymean, mymean], 'k')
```

### Figure 1: As the sample size grows, the average mean approaches the determined population mean (450) in the above Chi-square distrubition. This is depicted visually, as the plotted points gather more tightly as the sample size increases.

## NORMAL DISTRIBUTION

```python
#Setting the seed, the dictionary key, and key variables
np.random.seed(100)
mydict['norm'] = {} 
loc = mymean
scale = mySD

#Filling the dictionary
for n in mysize:
    mydict['norm'][n] = np.random.normal(loc, scale, size = n)

#Saving the means in my dictionary
mydict['norm']['means'] = {}

#Plotting the Means
axNorm_means = pl.figure(figsize=(15,8)).add_subplot(111)
for nn in mydict['norm']:
    if not type(nn) == str:
        mydict['norm']['means'][nn] = mydict['norm'][nn].mean()
        axNorm_means.plot(nn, mydict['norm']['means'][nn], 'o')
        axNorm_means.set_xlabel('sample size', fontsize=18)
        axNorm_means.set_ylabel('sample mean', fontsize=18)
        axNorm_means.set_title('Normal', fontsize=18)
        axNorm_means.plot([min(mysize), max(mysize)], [mymean, mymean], 'k')
```

### Figure 2: The above Normal distribution further showcases the Central Limit Theorum. As our sample size grows, the means approach the population mean

## POISSON DISTRIBUTION

```python
#Setting the seed, the dictionary key, and key variables
np.random.seed(150)
mydict['pois'] = {} 
lam = mymean

#Filling the dictionary
for n in mysize:
    mydict['pois'][n] = np.random.poisson(lam, size = n)

#Saving the means in my dictionary
mydict['pois']['means'] = {}

#Plotting the Means
axPois_means = pl.figure(figsize=(15,8)).add_subplot(111)
for nn in mydict['pois']:
    if not type(nn) == str:
        mydict['pois']['means'][nn] = mydict['pois'][nn].mean()
        axPois_means.plot(nn, mydict['pois']['means'][nn], 'o')
        axPois_means.set_xlabel('sample size', fontsize=18)
        axPois_means.set_ylabel('sample mean', fontsize=18)
        axPois_means.set_title('Poisson', fontsize=18)
        axPois_means.plot([min(mysize), max(mysize)], [mymean, mymean], 'k')
```

### Figure 3: The above Poisson Distribution again showcases the reality of selecting larger sample sizes: you are more likely to approximate the population mean

## BINOMIAL DISTRIBUTION

```python
#Setting the seed, the dictionary key, and key variables
np.random.seed(200)
mydict['binom'] = {} 
trials = 750
prob = 0.6
#trials*prob = mymean
#750*0.6 = 450

#Filling the dictionary
for n in mysize:
    mydict['binom'][n] = np.random.binomial(trials, prob, size = n)

#Saving the means in my dictionary
mydict['binom']['means'] = {}

#Plotting the Means
axBinom_means = pl.figure(figsize=(15,8)).add_subplot(111)
for nn in mydict['binom']:
    if not type(nn) == str:
        mydict['binom']['means'][nn] = mydict['binom'][nn].mean()
        axBinom_means.plot(nn, mydict['binom']['means'][nn], 'o')
        axBinom_means.set_xlabel('sample size', fontsize=18)
        axBinom_means.set_ylabel('sample mean', fontsize=18)
        axBinom_means.set_title('Binomial', fontsize=18)
        axBinom_means.plot([min(mysize), max(mysize)], [mymean, mymean], 'k')
```

### Figure 4: Once again, we see that the Central Limit Theorem holds tightly 

## GUMBEL DISTRIBUTION

```python
#Setting the seed, the dictionary key, and key variables
np.random.seed(250)
mydict['gumb'] = {} 
gamma = 0.5772156649015328606065120900824024310421
scale = (mySD*sqrt(6))/np.pi
mode = mymean - ((scale)*gamma)

#Filling the dictionary
for n in mysize:
    mydict['gumb'][n] = np.random.gumbel(mode, scale, size = n)

#Saving the means in my dictionary
mydict['gumb']['means'] = {}

#Plotting the Means
axGumb_means = pl.figure(figsize=(15,8)).add_subplot(111)
for nn in mydict['gumb']:
    if not type(nn) == str:
        mydict['gumb']['means'][nn] = mydict['gumb'][nn].mean()
        axGumb_means.plot(nn, mydict['gumb']['means'][nn], 'o')
        axGumb_means.set_xlabel('sample size', fontsize=18)
        axGumb_means.set_ylabel('sample mean', fontsize=18)
        axGumb_means.set_title('Gumbel', fontsize=18)
        axGumb_means.plot([min(mysize), max(mysize)], [mymean, mymean], 'k')
        
#Below is the math I used to determine my sclale and mode, based on the wikipedia page for the Gumbel Distribution.
#I could not find an elegant means of calling Euler's constant (gamma), so I defined it myself.

#np.random.gumel takes the following arguments: (mode, scale, size = n)
#where
#mode = mymean - scale*gamma
#scale = mySD*sqrt(6) / pi
#and
#gamma = 0.5772156649015328606065120900824024310421

#Adjusting these formulas, I was able to generate the ones used above in my code
```

### Figure 5: The Gumbel distribution showcases the accuracy of the Central Limit Theorem

# ALL MEANS

```python
allmeans = list(mydict['chisq']['means'].values() + 
               mydict['norm']['means'].values() +
               mydict['pois']['means'].values() + 
               mydict['binom']['means'].values() +
               mydict['gumb']['means'].values()
               )

pl.figure(figsize=(15, 15))
pl.hist(allmeans,bins=50)
pl.xlabel('sample mean', fontsize = 15)
pl.ylabel('Number of samples', fontsize = 15);
```

### Figure 6: The above histogram plots all 490 samples from the previous examples against their means. As evidenced by the density around our chosen population mean of 450, the Central Limit Theorum holds true.
