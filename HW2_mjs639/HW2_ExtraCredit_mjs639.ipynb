```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import os
from pandas import DataFrame, Series
from datetime import datetime
plt.style.use('seaborn-darkgrid')
```

```python
DFdata = os.getenv('DFDATA')
print(DFdata)
```

```python
#df = pd.read_csv(DFdata + "/kwmq-dbub/1414245703/kwmq-dbub")
df = pd.read_csv(DFdata + "/zkky-n5j3/1414246141/zkky-n5j3")
```

```python
df.head(5)
```

```python
dfRed = df[df.columns[0:2]]
dfRed = dfRed.rename(columns={'Neversink Date              ': 'NS Date', 'Neversink Elevation': 'NS Elev'})
dfRed['NS Date'] = pd.to_datetime(dfRed['NS Date'])
```

```python
dfRed.head(5)
```

```python
ax = dfRed.plot(x='NS Date', y='NS Elev', style='.', figsize=(15,8),title='Neversink Resevoir Elevation Levels')
ax.set_xlabel("Date")
ax.set_ylabel("Elevation Level")
print("Fig. 1 shows the elevation levels of the Neversink Reservoir over the span of several years")
```
