```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import os
from pandas import DataFrame, Series
from datetime import datetime
plt.style.use('seaborn-darkgrid')
```

```python
DFdata = os.getenv('DFDATA')
print(DFdata)
```

```python
df = pd.read_csv(DFdata + "/ia2d-e54m/1414245804/ia2d-e54m")
```

```python
df.head(5)
```

```python
dfRed = df.drop(['Per Capita(Gallons per person per day)'], axis=1, level=None, inplace=False, errors='raise')
```

```python
dfRed.head(5)
```

```python
ax = dfRed.plot.scatter(x='Year', y='NYC Consumption(Million gallons per day)', figsize=(12,8),
                title='Average Daily Water Consumption in NYC per year')
ax.set_xlabel("Year")
ax.set_ylabel("Avg. Daily Water Consumption (in millions)")

print('Fig. 1 is a plot of the average daily water consumption in NYC over the span of 30 years, beginning in 1979')
```
